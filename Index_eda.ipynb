```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("in_index.csv")
df['Dates']  = pd.to_datetime(df['Dates'])
print(df.shape)
df.tail(2)
```

```python
df33 = df[['Dates','Gold_price', 'AGG_price' , 'SP_price', 'HS_price', 'FTSE_price', 'DJ_price']].set_index('Dates')
df33 = df33[1:]
usrec2 = pd.read_csv('rec_data.csv')
usrec2 = usrec2[['Dates','USREC']]
usrec2['Dates'] = pd.to_datetime(usrec2.Dates)
usrec2.set_index('Dates', inplace = True)
usrec2 = usrec2[1:]
```

```python
df33.columns
```

```python
def plot_rec(x,y,t):
    sns.set()
    plt.figure(figsize = (20,10))
    plt.plot(x, y)
    plt.title( t + ' in Recession', fontsize = 30)
    plt.xlabel('Date', fontsize = 20)
    plt.ylabel(t, fontsize = 18)
    plt.xticks(fontsize = 15)
    plt.yticks(fontsize = 15)
    plt.fill_between(usrec2.index, min(y), max(y), where=usrec2['USREC'].values, color='gray', alpha=0.3)
    plt.xlim(x[1],x[-1])
```

```python
for i in df33.columns:
    plot_rec(df33.index, df33[i], i)
```

### Bottom in each recession

```python
def get_min(x,o,n):
    tt = df10[[x, 'USREC']]
    tt2 = tt[o:n].reset_index()
    mn = tt2[x].idxmin() 
    minn = tt2.at[mn, 'Dates']
    start = tt2.at[0, 'Dates']
    months = round((minn - start).days/ 30)
    return(months)
```

```python
### 1990 
for i in df10.columns.drop('USREC'):
    print('Bottom for ' + i +':',get_min(i,'1990-07-31','1991-02-28'))
print('Total length:', len(df10['1990-07-31':'1991-02-28']) , 'months')
```
